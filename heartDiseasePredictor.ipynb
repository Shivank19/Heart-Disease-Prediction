```python
#dependencies

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_df = pd.read_csv('heart.csv')
heart_df
```

```python
heart_df.shape
```

```python
#checking for null values
heart_df.isnull().sum()
```

```python
heart_df.describe()
```

```python
#splitting features and target
X = heart_df.drop(columns='target', axis=1)
y = heart_df['target']
```

```python
#train - test split
X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print ( X.shape, X_Test.shape, X_Train.shape)
```

```python
#Logistic Regression Model
model = LogisticRegression()

#traing LR model
model.fit(X_Train, y_Train)
```

## Model Evaluation

#### Metric - Accuracy Score

```python
#Accuracy on Training Data

X_Train_prediction = model.predict(X_Train)
train_data_accuracy = accuracy_score(X_Train_prediction, y_Train)
print('Accuracy on training data: ', train_data_accuracy)
```

```python
#Accuracy on Testing Data

X_Test_prediction = model.predict(X_Test)
test_data_accuracy = accuracy_score(X_Test_prediction, y_Test)
print('Accuracy on testing data: ', test_data_accuracy)
```

## Prediction Model

```python
input_data = (62, 0, 0, 140, 268, 0, 0, 160, 0, 3.6, 0, 2, 2)

input_data_np_Array = np.asarray(input_data)

#reshaping: so model predicts on one datapoint
input_data_reshaped = input_data_np_Array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
    print("Person does not have a Heart Disease")
else:
    print("Person is at risk of Heart Disease")
```
